```python
import pandas as pd
```

```python
# Load csv file

data= pd.read_csv('data/raw_international_debt.csv')
data
```

```python
data.info()
```

```python
# Looking into data for null values

data.isnull().sum()
```

```python
data.columns
```

```python
# Renaming columns for database
data = data.rename(columns={'Unnamed: 0': 'country_id',
                            'Country Name': 'country_name',
                           'Country Code':'country_code',
                           'Indicator Name': 'indicator_name',
                           'Indicator Code': 'indicator_code'})
data.head(2)
```

```python
data = data.drop(['2020','2021','2022', '2023', '2024', '2025', '2026', 'Unnamed: 61'], axis = 1) 
data.head(2)
```

```python
#Saving into file to use Tableau visualization
data.to_csv('data/tableau_international_debt.csv', index=False)
```
